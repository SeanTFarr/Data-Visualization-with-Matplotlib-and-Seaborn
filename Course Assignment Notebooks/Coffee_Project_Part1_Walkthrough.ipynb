# Coffee Production Stats

* Create a bar chart of the top 10 coffee producers by total production in 'total-production.csv', sorted from highest to lowest. Change the xticks to two letter country abbreviations.


* Data Prep Code

<mark style="background-color: #000000"> coffee_production = pd.read_csv("coffee/total-production.csv").T </mark> 

<mark style="background-color: #000000"> coffee_production.columns = coffee_production.iloc[0] </mark> 

<mark style="background-color: #000000"> coffee_production = coffee_production.drop("total_production") </mark> 

```python
path = "../Data/coffee/total-production.csv"

coffee_production=pd.read_csv(path)#.T # Adding ".T" reads it in transposed 

coffee_production.head()
```

```python
# transposing the data to get the years as our rows and the countries as our columns
coffee_production = coffee_production.T
```

```python
coffee_production.head()
```

```python
# Setting the column names to equalthe first row
coffee_production.columns = coffee_production.iloc[0]
```

```python
coffee_production.head()
```

```python
# Dropping the first row
coffee_production = coffee_production.drop("total_production")
```

```python
coffee_production.head()
```

```python
coffee_production.info()
```

```python
top10_producers = (coffee_production
                  .sum()# to get the sum of each column
                   .sort_values(ascending=False) # to get in order (sort_index would give alphabetical order)
                   .iloc[:10]) # pulling only the top 10

top10_producers
```

```python
fig,ax = plt.subplots()

# select data
ax.bar(top10_producers.index,top10_producers/1000000) 
# can also be written ax.bar(x=top10_producers.index,height=top10_producers/1000000)

# Setting up custom xticks and xticklabels
#create a range of integers
positions = list(range(10))
# attach labels to the ingtegers
labels = ["BR", "VN", "CO", "ID", "ET", "IN", "MX", "GT", "HN", "UG"]

# Adding the xticks and xticklabels
ax.set_xticks(positions)
ax.set_xticklabels(labels)

# set title and ylabel
ax.set_title("Top Coffee Producing Nations 1990-2018")
ax.set_ylabel("Production (Millions 60kg Bags)", fontsize=12)

#removing spines
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False);
```

## Composition Over Time

Our client is interested in looking at how Brazil's share of global coffee production has changed over time.

Build a plot that global total production as well as Brazil's contribution to global production over time.

```python
# Creating a "brazil_vs_others" DataFrame and a "rest_of_world" column using .assign
brazil_vs_others = (coffee_production
                   .assign(rest_of_world = coffee_production
                          .drop("Brazil", axis=1)# drop the Brazil data from the column
                           .sum(axis=1)) # summ up the columns
                   .loc[:,["Brazil", "rest_of_world"]] #grabbing the columns
                   .astype({"Brazil": "float64"})
                   )

brazil_vs_others.dtypes
```

```python
brazil_vs_others.head()
```

```python
fig, ax = plt.subplots()

labels=["Brazil", "World Total"]
#colors=["blue", "orange"]

ax.stackplot(
    brazil_vs_others.index.astype("datetime64[ns]"), #must cast as datetime64 for clear x axis titles
    brazil_vs_others["Brazil"],
    brazil_vs_others["rest_of_world"],
    labels=labels,
#    colors=colors
)

# set figure title, plot title, and y axis label
fig.suptitle("Brazil's Share of Global Coffe Production 1990-2018")
ax.set_title("Brazil Increases Share of Global Market")
ax.set_ylabel("Production (60kg Bags)", fontsize=12)

#removing spines
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# call legend and set placement
ax.legend(loc="upper left")
```

# Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries. 

Is there a relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
# Read in imports 
imports = pd.read_csv("../Data/coffee/imports.csv")

imports.head()
```

```python
# To clean it up
# Set the index to be the 'Imports' column
# grabbing the mean of the columns
consumption = pd.read_csv("../Data/coffee/imports.csv").set_index("imports").mean(axis=1)

consumption.name= "imports"

consumption.head()
```

```python
# Read in retail-prices
prices = pd.read_csv("../Data/coffee/retail-prices.csv")

prices.head()
```

```python
prices = pd.read_csv("../Data/coffee/retail-prices.csv").set_index("retail_prices").mean(axis=1)

prices.head()
```

```python
# Merge the two 
price_cons =(pd.DataFrame(prices)
            .merge(consumption,
                  left_on=prices.index,
                  right_on=consumption.index,
                  how="inner" #leaves only countries in both dataframes
                )
            .sort_values(["imports"], ascending = False)
)
# renaming columns
price_cons.columns=["Country","Price", "Imports"]
# strip away leading white spaces in Country column
price_cons["Country"]=price_cons["Country"].str.strip(" ")

price_cons
```

```python
fig,ax=plt.subplots(figsize=(14,10))

# add a width and a length of index
width=.35
x=np.arange(0,len(price_cons["Country"]))

# first bar
bar1 = ax.bar(x-width/2,
              price_cons["Imports"],
              width=width,
              label="Price")
ax2 = ax.twinx()

# second bar
bar2 =ax2.bar(x+width/2,
              price_cons["Price"],
              width=width,
              label="Imports",
              color="orange")


ax.set_title("Prices Paid Vs. Consumption of Select Importing Nations", fontsize=18, fontweight="bold")
ax.set_ylabel("Consumption (60k Bags)", fontsize=12)
ax2.set_ylabel("Average Price Paid Per Bag(USD)", fontsize=12)

#use plt.xticks
plt.xticks(x, price_cons["Country"], fontsize=7)

ax.legend([bar1, bar2], ["Comsumption", "Price"]);
```

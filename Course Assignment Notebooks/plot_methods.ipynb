```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

### The basics

```python
data = [x * 10 for x in range(10)]

data
```

```python
labels = [f'{x}!' for x in range(10)]

labels
```

```python
plt.plot(data)
```

```python
plt.plot(labels, data)
```

```python
# to get rid of the extra line of text at the top of the chart, under the code
# you can either place the ";" at the end of the code, or add the line "plt.show()"
plt.plot(labels, data)

plt.show()
```

```python
# creating a dataframe from the current variables and adding a new column using ".assign"
# "data" is our value column, "labels" is set  as the index
df = pd.DataFrame(data, labels).assign(series2 = lambda x: x[0] / 2)

df.head()
```

```python
plt.plot(df);
```

### Plotting Methods

```python
# objest oriented approach - fig and ax set up the chart
fig = plt.figure()

ax = fig.add_subplot()
```

```python
# objest oriented approach - plotting "data" and adding titles
fig = plt.figure()

ax = fig.add_subplot()

ax.plot(data)

ax.set_title("Title")
fig.suptitle("Overall Title");
```

```python
# fig and ax on same line of code
fig, ax = plt.subplots()

ax.plot(data)

ax.set_title("Title")
fig.suptitle("Overall Title");
```

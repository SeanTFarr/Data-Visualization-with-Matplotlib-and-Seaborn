```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_raw = pd.read_csv(
    "../Data/housing_data.csv",
    parse_dates=["period_begin", "period_end"],
dtype={"total_homes_sold": "Int32", "region_name": "category"})
```

```python
housing_raw.head()
```

```python
housing_raw.info()
```

```python
ca_housing = housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
].assign(region_name = housing_raw["region_name"].str[0:-11])
```

```python
ca_housing.info(memory_usage="deep")
```

```python
ca_housing.head()
```

```python
ca_housing_pivot= ca_housing.pivot_table(index="period_begin",
                                        columns = "region_name",
                                        values= "median_active_list_price").assign(
                                        CA_average= lambda x: x.mean(axis=1))
ca_housing_pivot.head()
```

```python
#narrowing down which regions we are lookng at
ca_housing_markets= ca_housing_pivot.loc[:, ["San Francisco",
                                            "Los Angeles",
                                            "San Diego",
                                            "CA_average"]]
ca_housing_markets.head()
```

```python
fig, ax = plt.subplots()

#ax.plot(ca_housing_markets)

ax.plot(ca_housing_markets["San Francisco"])
ax.plot(ca_housing_markets["Los Angeles"])

ax.set_title("Housing Prices in CA markets 2017-2022", fontsize=14)
#fig.suptitle("");
ax.set_xlabel("Year")
ax.set_ylabel("Price (Millions)")



plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,6))

ax.plot(
        ca_housing_markets.index, 
        ca_housing_markets["San Francisco"]/ 1000000, 
        label="San Francisco",
        color="purple",
        linewidth=2
)
ax.plot(
        ca_housing_markets.index, 
        ca_housing_markets["Los Angeles"]/ 1000000,
        label="Los Angeles",
        color="blue",
        linewidth=2,
        linestyle="--"
)
ax.set_xlabel("Year")
ax.set_ylabel("Price (Millions)")
ax.set_ylim(0,1.6)
ax.axvline(18341, c="black", ls="--", label="important date")
# annotation instead of text
ax.annotate("Start of Covid",
           xy=(18300,1.3),
           xytext=(17600,1.2),
           arrowprops=dict(
               facecolor="black",
               width=.5,
               headwidth=6,
               connectionstyle="angle3, angleA=290, angleB=0"
           ),
           verticalalignment="center")
#removing spines
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)


ax.legend()

plt.show()
```

```python
# by listing colors and linestyles then running forloop
colors= ["orange", "grey", "grey", "grey"]
ls=["-", "--",":", "-."]

fig, ax = plt.subplots(figsize=(8,6))

for i, column in enumerate(ca_housing_markets.columns):
    ax.plot(
        ca_housing_markets[column]/1000000,
        color=colors[i],
        ls=ls[i])

ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price")
ax.set_ylim(0,1.6)
ax.axvline(18341, c="black", ls="--", #label="Start of Covid -->"
          )
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.legend(ca_housing_markets.columns, loc="lower left")

ax.annotate("Start of Covid",
           xy=(18300,1.3),
           xytext=(17600,1.2),
           arrowprops=dict(
               facecolor="black",
               width=.5,
               headwidth=6,
               connectionstyle="angle3, angleA=290, angleB=0"
           ),
           verticalalignment="center")
fig.text(.15,
        .9,
        "San Francisco Prices Hit Hard By Covid...",
        fontsize=14
        )
ax.text(18400, 1,
       "..while prices elsewhere\nin California kept rising",
       fontsize=12)

plt.show()
```

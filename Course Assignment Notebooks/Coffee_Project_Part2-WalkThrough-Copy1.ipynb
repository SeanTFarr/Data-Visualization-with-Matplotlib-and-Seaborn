```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
coffee_production = pd.read_csv("../data/coffee/total-production.csv").T

coffee_production.columns= coffee_production.iloc[0]

coffee_production.drop("total_production", inplace=True)
```

```python
# setting axes parameter settings
axes_settings = {"spines.top":False,
                "spines.right":False,
                }

plt.rc("axes", **axes_settings)
```

### Data Prep

```python
# Creating a "brazil_vs_others" DataFrame and a "rest_of_world" column using .assign
brazil_vs_others = (coffee_production
                   .assign(rest_of_world = coffee_production
                          .drop("Brazil", axis=1)# drop the Brazil data from the column
                           .sum(axis=1)) # summ up the columns
                   .loc[:,["Brazil", "rest_of_world"]] #grabbing the columns
                   .astype({"Brazil": "float64"})
                   )

brazil_vs_others.head()
```

```python
# to create rest of world column(could also use assign, like above) 
coffee_production["Rest of World"] = coffee_production.drop(["Brazil",
                                                            "Viet Nam",
                                                            "Colombia",
                                                            "Indonesia",
                                                            "Ethiopia",], axis=1).sum(axis=1)
# make a series to hold top 5 and rest of world and grab the row for 2018 
coffee_production_top5_2018=coffee_production.loc['2018', ["Brazil",
                                                          "Viet Nam",
                                                           "Colombia",
                                                           "Indonesia",
                                                           "Ethiopia",
                                                           "Rest of World"]
                                                 ]


# to create dataframe
pd.DataFrame(coffee_production_top5_2018)
```

```python
#set colors
colors=["#009639","#C8102E", "#FFCD00", "#EF3340", "#009A44", "grey"]
# pie chart,needs section exploded and colors
fig,ax=plt.subplots()

ax.pie(x=coffee_production_top5_2018[::-1],
      startangle=90,
      labels=["Brazil", "Viet Nam", "Colombia", "Indonesia", "Ethiopia", "Rest of World"][::-1],
      autopct="%.0f%%",
       colors=colors[::-1],
       explode=(0,0,0,0,0,.1)
      )
ax.set_title("Share of Global Coffee Production 2018");
```

```python
fig,ax = plt.subplots()

# select data
ax.barh(coffee_production_top5_2018.index,
       (coffee_production_top5_2018),
      color=colors)
#revers the order of the bars
plt.gca().invert_yaxis()

ax.set_title("Top Coffee Producing Nations 1990-2018")
ax.set_xlabel("Production (Millions 60kg Bags)")
```

```python
# donut for 1990
fig, ax = plt.subplots()
# Pull 1990 data (first row) using .iloc, then sort the values
ax.pie(brazil_vs_others.iloc[0].sort_values(ascending = False),
      startangle=90,
      labels=["",""],
#       autopct="%.0f%%",
       pctdistance=.85,
      colors=["white","#009639"]
)

# to create circle
hole= plt.Circle((0,0), 0.7, fc="white")
fig = plt.gcf()
# adding the circle in the pie chart
fig.gca().add_artist(hole)

# to add center text
plt.text(0,
        0,f"{round((brazil_vs_others.loc['1990','Brazil'] / brazil_vs_others.loc['1990'].sum()*100))}%", 
        ha="center",
        va="center",
        fontsize=42
         
)

ax.set_title("Brazil Share of Global Production 1990", fontsize=12);
```

```python
# donut for 2018
fig, ax = plt.subplots()

# Pull 2018 data (Last row) using .iloc, then sort the values
ax.pie(brazil_vs_others.iloc[-1].sort_values(ascending = False),
      startangle=90,
      labels=["",""],
#       autopct="%.0f%%",
#       pctdistance=.85,
      colors=["white","#009639"]
)

# to create circle
hole= plt.Circle((0,0), 0.7, fc="white")
fig = plt.gcf()
# adding the circle in the pie chart
fig.gca().add_artist(hole)

# to add center text
plt.text(0,
        0,
        f"{round((brazil_vs_others.loc['2018', 'Brazil'] / brazil_vs_others.loc['2018'].sum()*100))}%",
        ha="center",
        va="center",
        fontsize=42
)

ax.set_title("Brazil Share of Global Production 2018", fontsize=12);
```

```python
# Stacked line chart
fig, ax = plt.subplots()

labels=["Brazil", "World Total"]
color=["#009639", "grey"]

ax.stackplot(
    brazil_vs_others.index.astype("datetime64[ns]"), #must cast as datetime64 for clear x axis titles
    brazil_vs_others["Brazil"],
    brazil_vs_others["rest_of_world"],
    labels=labels,
    colors=color
)

#  plot title, and y axis label
ax.set_title("Brazil's Share of Market Grows as Global Output Surges")

ax.set_ylabel("Production (60kg Bags)", fontsize=12)

#removing spines
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# call legend and set placement
ax.legend(loc="upper left");
```

### Meshgrid Layout

```python
from matplotlib.gridspec import GridSpec
# set the figure
fig = plt.figure(figsize=(11, 8.5))

#main title
fig.suptitle("Brazil Extends it's Dominance in the Growing Coffee Market",
             fontsize=16,
            fontweight="bold")

#set up the grid
grid = GridSpec(12,6, figure=fig)

#set colors
colors=["#009639","#C8102E", "#FFCD00", "#EF3340", "#009A44", "grey"]

# Text "chart"
# ax1 = fig.add_subplot(grid[0:3, 0:4]) was wrong about this one
#use fig.text, starting with x,y location
fig.text(.05, .745,
'''Brazil is the world's top coffee grower.



Since 1990,Brazil's share of global
productinhas risen from 29% to37%,
even as the  worldwide market has
expanded by over 50%'''  ,
fontsize=12         
)

#1st donut
ax2 = fig.add_subplot(grid[0:4, 2:4])
# Pull 1990 data (first row) using .iloc, then sort the values
ax2.pie(brazil_vs_others.iloc[0].sort_values(ascending = False),
      startangle=90,
      labels=["",""],
#       autopct="%.0f%%",
       pctdistance=.85,
      colors=["white","#009639"]
)

# to create circle
hole= plt.Circle((0,0), 0.7, fc="white")
fig = plt.gcf()
# adding the circle in the pie chart
fig.gca().add_artist(hole)

# to add center text
plt.text(0,
        0,f"{round((brazil_vs_others.loc['1990','Brazil'] / brazil_vs_others.loc['1990'].sum()*100))}%", 
        ha="center",
        va="center",
        fontsize=25
         
)

ax2.set_title("Brazil % Global Production 1990", fontsize=12)

#2nd donut
ax3 = fig.add_subplot(grid[0:4, 4:])
# Pull 2018 data (Last row) using .iloc, then sort the values
ax3.pie(brazil_vs_others.iloc[-1].sort_values(ascending = False),
      startangle=90,
      labels=["",""],
#       autopct="%.0f%%",
#       pctdistance=.85,
      colors=["white","#009639"]
)

# to create circle
hole= plt.Circle((0,0), 0.7, fc="white")
fig = plt.gcf()
# adding the circle in the pie chart
fig.gca().add_artist(hole)

# to add center text
plt.text(0,
        0,
        f"{round((brazil_vs_others.loc['2018', 'Brazil'] / brazil_vs_others.loc['2018'].sum()*100))}%",
        ha="center",
        va="center",
        fontsize=25
)

ax3.set_title("Brazil % Global Production 2018", fontsize=12)

# Stacked Line
ax4 = fig.add_subplot(grid[4:7, :])
labels=["Brazil", "World Total"]
color=["#009639", "grey"]

ax4.stackplot(
    brazil_vs_others.index.astype("datetime64[ns]"), #must cast as datetime64 for clear x axis titles
    brazil_vs_others["Brazil"],
    brazil_vs_others["rest_of_world"],
    labels=labels,
    colors=color
)

#  plot title, and y axis label
ax4.set_title("Brazil's Share of Market Grows as Global Output Surges")

ax4.set_ylabel("Production (60kg Bags)", fontsize=12)

# call legend and set placement
ax4.legend(loc="upper left")

# BarH
ax5 = fig.add_subplot(grid[8:, 0:2])
ax5.barh(coffee_production_top5_2018.index,
       (coffee_production_top5_2018),
      color=colors)
#revers the order of the bars
plt.gca().invert_yaxis()

ax5.set_title("Top Coffee Producing Nations 1990-2018")
ax5.set_xlabel("Production (Millions 60kg Bags)")
# Pie
ax6 = fig.add_subplot(grid[8:, 2:])
ax6.pie(x=coffee_production_top5_2018[::-1],
      startangle=90,
      labels=["Brazil", "Viet Nam", "Colombia", "Indonesia", "Ethiopia", "Rest of World"][::-1],
      autopct="%.0f%%",
       colors=colors[::-1],
       explode=(0,0,0,0,0,.1)
      )
ax6.set_title("Share of Global Coffee Production 2018")

plt.show()
```

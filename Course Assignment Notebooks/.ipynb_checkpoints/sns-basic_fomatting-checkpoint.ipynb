## Seaborn Formatting Basics

```python
import pandas as pd
import seaborn as sns
```

```python
housing_raw = pd.read_csv(
    "../Data/housing_data.csv",
    parse_dates=["period_begin", "period_end"],
dtype={"total_homes_sold": "Int32", "region_name": "category"})

ca_housing = (housing_raw
    .loc[(housing_raw["region_name"]
    .str.contains("CA"))]
    #deleting the last 11 spaces in "region_name"(i.e."County,GA")
    .assign(region_name = housing_raw["region_name"].str[0:-11])
    #.set_index("period_begin").sort_index() # setting different index and sorting it
)

ca_housing_pivot= ca_housing.pivot_table(index="period_begin",
                                        columns = "region_name",
                                        values= "median_active_list_price").loc[:, 
                                                                                ["San Francisco",
                                                                                 "Los Angeles",
                                                                                 "San Diego",
                                                                                 "Tulare"]].assign(overall_avg= lambda x: x.mean(axis=1))
ca_housing_pivot.head()
```

```python
#running on an aggregated dataframe(pivot)
sns.lineplot(
    data=ca_housing_pivot,
    x=ca_housing_pivot.index,
    y="overall_avg"
)
sns.despine()
```

```python
# a non aggregated dataframe
ca_housing.head()
```

```python
sns.lineplot(
    x="period_begin",
    y="median_active_list_price",
    data=ca_housing,
#     ci=None, #to supress the confidence intervals
#     estimator=min # changes the default(mean)
)

sns.despine()

# this gives the average price with error bars,based off of the standard deviations
# (by default a 95% confidence interval, 2 stand dev away from mean )
```

```python
sns.lineplot(
    x="period_begin",
    y="median_active_list_price",
    data=ca_housing,
    ci=None, #to supress the confidence intervals
    estimator=min # changes the default(mean)
)

sns.despine()
```

```python
sns.lineplot(
    x="period_begin",
    y="median_active_list_price",
    data=ca_housing,
    errorbar=None, #to supress the confidence intervals
    estimator=min # changes the default(mean)
)

sns.despine()
```

```python
sns.lineplot(
    x="period_begin",
    y="median_active_list_price",
    data=ca_housing.query("region_name in ['Los Angeles', 'San Francisco']"),
    errorbar=None,
    hue="region_name",
    palette="husl",
#    ls='--'
)

sns.despine()
```

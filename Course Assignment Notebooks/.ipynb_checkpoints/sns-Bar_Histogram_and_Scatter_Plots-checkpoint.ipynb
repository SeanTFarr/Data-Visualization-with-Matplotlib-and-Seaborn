## Seaborn Bar Charts

```python
import pandas as pd
import seaborn as sns
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv").drop(["Unnamed: 0"], axis=1)

diamonds.head()
```

### Looking at "cut" and "carat"

```python
sns.barplot(
    x="cut",
    y="carat",
    data=diamonds,
    #errorbar=None
)
# black bars indicate aggregation being performed
```

```python
sns.barplot(
    x="carat",
    y="cut",
    data=diamonds,
    errorbar=None
)
# by switching the numerical and category columns, you get a horizontal bar chart
```

```python
#grouped bar charts
sns.barplot(
    x="carat",
    y="cut",
    hue="color",
    data=diamonds.sort_values(by="color"),
    palette="viridis"
) #'hue' allows us to further split the data into additional categories
```

```python
#histogram
sns.histplot(x="price", data=diamonds)
```

```python
# you can specify the number of "bins" and add kernel density(kde=True)
sns.histplot(x="price",
             bins=10,
             kde=True,
             data=diamonds)
```

### Looking at "cut" and "price"

```python
#second demo comparing "cut" & "price"
sns.barplot(
    x="cut",
    y="price",
    data=diamonds,
    #errorbar=None
)
```

```python
sns.barplot(
    x="price",
    y="cut",
    data=diamonds,
    errorbar=None
)
```

```python
sns.barplot(
    x="price",
    y="cut",
    data=diamonds,
    facecolor="white",
    edgecolor="black"
    
)
```

```python
# adding "hue"
sns.barplot(
    x="cut",
    y="price",
    hue="clarity",
    data=diamonds,
    #errorbar=None
)
```

```python
# To modify theabove chart, we can use matplotlib
import matplotlib.pyplot as plt
#increase the figure size
fig, ax = plt.subplots(figsize=(8, 8))

sns.barplot(
    x="cut",
    y="price",
    hue="clarity",
    data=diamonds,
    #errorbar=None
)
#move the legend outside of the chart
ax.legend(bbox_to_anchor=(1.2,1))
```

### Histograms

```python
#using .query to select two to compare
sns.histplot(
    x="price",
    hue="clarity",
    data=diamonds.query("clarity in ['I1', 'IF']"),
    palette="hls",
    stat="percent", #to look at percent
#    bins=10,
    binwidth=1000,
    kde=True
)
```

```python
#
sns.histplot(
    x="price",
#     y="clarity",
    data=diamonds,
#     cbar=True
)
```

```python
#heatmap view by adding in the second categorical variable(density plot)
sns.histplot(
    x="price",
    y="clarity",
    data=diamonds,
#     cbar=True
)
```

```python
# Adding the cbar give a gauge for the values
sns.histplot(
    x="price",
    y="clarity",
    data=diamonds,
    cbar=True
)
```

### Scatter Plots

```python
sns.scatterplot(
    x="x",
    y="price",
    data=diamonds,
    hue="carat",
    palette="husl",
    size="carat"
)
```

```python
sns.scatterplot(
    x="price",
    y="carat",
    data=diamonds,
    hue="clarity",
    palette="husl",
#     size="carat"
)
```

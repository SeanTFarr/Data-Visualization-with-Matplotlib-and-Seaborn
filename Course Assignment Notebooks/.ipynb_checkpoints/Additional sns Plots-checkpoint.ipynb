## Boxplots, Box & Violin, and Linear Relationship Charts

### Boxplots

```python
import pandas as pd
import seaborn as sns
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv").drop(["Unnamed: 0"], axis=1)

diamonds.head()
```

```python
sns.boxplot(x="price",
           data= diamonds)
```

```python
sns.boxplot(x="cut",
            y="price",
           data= diamonds)
```

```python
sns.boxplot(x="clarity",
            y="price",
           data= diamonds)
```

### Violin Plots

```python
#Violin Plots show the KDE curve on the sides
sns.violinplot(x="cut", 
               y="price", 
               data= diamonds)
```

```python
sns.violinplot(y="price", 
               data= diamonds)
```

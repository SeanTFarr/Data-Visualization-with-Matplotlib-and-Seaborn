## Colors

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating data, integers 0-9
data = [x for x in range(10)]
# integers 2 times 0-9
data2 = [x * 2 for x in range(10)]
```

```python
# changing color parameters

#with specified colors
#plt.rcParams["axes.prop_cycle"]= plt.cycler(color=["red", "orange", "grey", "green"])

#With preset colormap
plt.rcParams["axes.prop_cycle"]= plt.cycler("color", plt.cm.Set1.colors)
```

```python
fig, ax= plt.subplots()

ax.plot(data,
      #c="green"
      )
ax.plot(data2,
      #c="blue"
      )
```

## Style Sheets

```python
#set style sheet (will work on all once set)
plt.style.use("fivethirtyeight")
```

```python
fig, ax= plt.subplots()

ax.plot(data,
      #c="green"
      )
ax.plot(data2,
      #c="blue"
      )
```

## Parameters
Runtime Configuration Parameters

```python
plt.rc("axes.spines", top=False, right=False)

#set linestyle (will work on all once set)
plt.rcParams["lines.linestyle"] = "--"
```

```python
fig, ax = plt.subplots()

ax.plot(data)
```
